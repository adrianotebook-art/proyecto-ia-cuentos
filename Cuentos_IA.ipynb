```python
# Generador de cuentos infantiles con IA

def generar_prompt_cuento(tema):
    prompt = f"""
    Actúa como un escritor profesional de cuentos infantiles.

    Escribe un cuento corto para niños de 6 años.
    Tema: {tema}

    Requisitos:
    - Lenguaje simple
    - Máximo 200 palabras
    - Incluir una enseñanza final
    """
    return prompt

tema = "un dragón que aprende a compartir"

prompt = generar_prompt_cuento(tema)

print("PROMPT GENERADO:\n")
print(prompt)
```

```python
cuento = """Había una vez un pequeño dragón llamado Lumo. Lumo tenía una cueva llena de tesoros: monedas brillantes, piedras de colores y dulces que le encantaban. Pero había un problema: no le gustaba compartir nada.

Un día, una ardilla llamada Nia llegó a su cueva. Tenía hambre y frío.
—¿Puedes compartir un dulce conmigo? —preguntó.

Lumo dudó. Miró sus dulces… y dijo que no.

Nia se fue triste. Esa noche, Lumo se sintió solo. Nadie venía a jugar con él.

Al día siguiente, Lumo decidió buscar a Nia. La encontró bajo un árbol.
—Lo siento —dijo—. ¿Quieres compartir mis dulces?

Nia sonrió. Compartieron dulces, risas y juegos. Pronto, más animales se unieron. La cueva de Lumo se llenó de amigos.

Lumo descubrió algo importante: compartir no le quitaba nada, sino que le daba mucho más.

Enseñanza: Compartir nos ayuda a tener amigos y ser más felices."""

print(cuento)
```
